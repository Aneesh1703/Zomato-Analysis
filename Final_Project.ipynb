```python
a = pd.read_excel('Country-Code.xlsx')
a.head()
```

```python
b = pd.merge(df,a, on="Country Code", how="left")
b.head()
```

```python
v1=b.Country.value_counts().values
v1
```

```python
v2=b.Country.value_counts().index
v2
```

```python
plt.pie(v1[:3],labels=v2[:3], autopct='%1.2f%%')
plt.show()
```

```python
b.head()
```

```python
b['Aggregate rating'].value_counts()
```

```python
ratings = b.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
ratings.head()
```

```python
fig = plt.figure(figsize=(12,3), dpi=100)
ax = fig.add_axes([1,1,1,2])
sns.barplot(x='Aggregate rating',y='Rating Count',data = ratings)
```

```python
fig = plt.figure(figsize=(12,6), dpi=100)
ax = fig.add_axes([2,2,1,1])
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
sns.countplot(x = "Rating color", data = ratings, palette=
              ['blue','red','orange','yellow','green','green'])
```
